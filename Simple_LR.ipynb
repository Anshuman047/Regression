## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Import Datasets


```python
#Importing Dataset from pandas which will create the dataframe of the datasets
dataset=pd.read_csv("Salary_Data.csv")
```

## Segregating the features and depenedent variables

```python
X=dataset.iloc[:,:-1].values # ":-1" means all the coloumns expect the last dependentColumn
Y=dataset.iloc[:,-1].values # "-1" means the last column byDefault
#.values because to get the datasets in the form of numpyArray
```

### (*) Hence no missing value is there, therefore not operating that Nan_processing
### (*) No strnigs are there, therefore no operating that Encodeing_processing

## Splitting the dataset into training and testing sets


```python
# Splitting the dataset into Training and testing Dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,random_state=2)
#Splitted into 80-20 % and when this cell run, the splitting would be always same as random_state=1
```

### (*) Therefore no multiple coloumns/features with different distribution capacity, so no FEATURE_SCALING required and also here it is not required because the coefficint would manage the capacity of the features

## Now importing the model and train it

```python
#Importing the model and trainig it
from sklearn.linear_model import LinearRegression
regressor= LinearRegression() #Build the instances of the LR model
regressor.fit(X_train,Y_train) #Training of the imported model from the training_datasets
```

## Predicting the Test_results

```python
Y_pred= regressor.predict(X_test)
```

## Visualizing the on predicted_Training sets

```python
# Visualising using pyplot module of matplot_library
plt.scatter(X_train,Y_train,c="r") #PLotting the datasets in redPoints
plt.plot(X_train,regressor.predict(X_train)) #Plotting the RegressorLine on trainingSets
plt.title("Salary VS Experience (Training_set)") #Titling of the Graph
plt.xlabel("Years of Experience") #Titling of X-axis
plt.ylabel("Amount of Salary") #Titling of Y-axis
plt.show
```

## Visualizing the on predicted_Testing sets

```python
# Visualising using pyplot module of matplot_library
plt.scatter(X_test,Y_test,c="r") #PLotting the datasets in redPoints
plt.plot(X_test,Y_pred) #Plotting the RegressorLine on trainingSets
plt.title("Salary VS Experience (Test_set)") #Titling of the Graph
plt.xlabel("Years of Experience") #Titling of X-axis
plt.ylabel("Amount of Salary") #Titling of Y-axis
plt.show
```

## Making a single prediction (for example the salary of an employee with 12 years of experience)

```python
print(regressor.predict([[12]]))
```

## (FOR ABOVE) Therefore, our model predicts that the salary of an employee with 12 years of experience is $ 139228.

Important note: Notice that the value of the feature (12 years) was input in a double pair of square brackets. That's because the "predict" method always expects a 2D array as the format of its inputs. And putting 12 into a double pair of square brackets makes the input exactly a 2D array. Simply put:

12→scalar

[12]→1D array

[[12]]→2D array

## Getting the final linear regression equation with the values of the coefficients

```python
print(regressor.coef_)
print(regressor.intercept_)
```
