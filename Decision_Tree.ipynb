```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Importing datasets
dataset=pd.read_csv("Position_Salaries.csv")
```

```python
# Getting our X and Y
X=dataset.iloc[:,1:-1].values
Y=dataset.iloc[:,-1].values
```

## From the Tree module, which contains DecisionTree-class is been imported, and training upon the datasets
### Recomended not to spend too much time in tunning the parameters here

```python
from sklearn.tree import DecisionTreeRegressor
# ObjectCreation
regressor=DecisionTreeRegressor(random_state=0) #There are various randomness in it, for having same scnario with course, both applied this for ending up at same state
regressor.fit(X,Y)
```

```python
Y_pred=regressor.predict([[6.5]])
print(Y_pred)
```

## Visualisation of the model high resolution {more smoothNess}, means lineFunction drawn over also decimalFeatures for getting perfectCurve

```python
# So converted the X into decimal points from integer points for more inputs and outputs, which leds the line smooth
X_grid=np.arange(min(X),max(X)+0.1,0.1) #Arranging from minimum of X to maximum of X with difference 0.1
X_grid1=X_grid.reshape(len(X_grid),1) #Same meaning, to convert the horizontal array into verticle, means column will become 1 from len(X) basically become transpose
X_grid2=X_grid.reshape(-1,1) #Same meaning as above but from CN procedure
print(X_grid)
```

```python
plt.scatter(X,Y,c="b")
plt.plot(X_grid1,regressor.predict(X_grid1),c="r")
plt.title("MorePrecised curve of DecisionTree_Regression")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(X,Y,c="b")
plt.plot(X,regressor.predict(X),c="r")
plt.title("Normal curve of DecisionTree_Regression")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```
